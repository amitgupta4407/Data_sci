```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# tittanic
# importing titanic data set
import requests
from io import StringIO
url = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/84196db26ed0a06ba55bd7ba44a7a8aae82ffdf4/day39-knn-imputer/train.csv"
headres ={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
# add header file to avoid 403
req = requests.get(url, headers=headres)
data = StringIO(req.text)
titanic = pd.read_csv(data)
titanic.head()
```

```python
flights = sns.load_dataset('flights')
flights
```

# 2. Bar plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(titanic["Pclass"],titanic["Age"]) #Black bar represent confidence interval
```

```python
sns.barplot(titanic["Pclass"],titanic["Fare"],hue=titanic["Sex"]) 
```

# 3. Box plot (Numerical - Categorical)

```python
sns.boxplot(titanic["Sex"], titanic["Age"],hue=titanic["Survived"])
```

# 4. Dist plot (Numerical - Categorical)

```python
sns.distplot(titanic["Age"])
```

```python
sns.distplot(titanic[titanic["Survived"]==0]["Age"],hist = False) #Dist plot of person's age not survived
sns.distplot(titanic[titanic["Survived"]==1]["Age"],hist = False, color = "red") #Dist plot of person's age survived
```

# 5. Heat Map (Categorical - Categorical)

```python
# har Pclass me kitna log mar gaya aur ktna bach gaya
pd.crosstab(titanic["Pclass"],titanic["Survived"])
```

```python
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic["Survived"]))
```

```python
(titanic.groupby('Pclass').mean()["Survived"]*100)   #this is scaler
```

```python
# (titanic.groupby('Pclass').mean()["Survived"]*100).plot()
(titanic.groupby('Pclass').mean()["Survived"]*100).plot(kind = "bar")
```

```python
(titanic.groupby('Sex').mean()["Survived"]*100).plot(kind = "bar")
```

6. Cluster Map (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic["Parch"],titanic["Survived"]))
```
