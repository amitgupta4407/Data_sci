```python
import pandas as pd
import seaborn as sns
```

```python
# importing titanic data set
import requests
from io import StringIO
url = "https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/84196db26ed0a06ba55bd7ba44a7a8aae82ffdf4/day39-knn-imputer/train.csv"
headres ={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
# add header file to avoid 403
req = requests.get(url, headers=headres)
data = StringIO(req.text)
df = pd.read_csv(data)
df.head()
```

# 1. Categorical Data

a. Countplot

```python
#sns.countplot(df["Survived"])
df["Survived"].value_counts().plot(kind = "bar")
```
